```python
# Create content_corpus.md file
content_corpus = """# Emergency First Aid Content Database

## CARDIAC EMERGENCIES

### 1. Cardiac Arrest - Unresponsiveness and No Pulse
**Signs:** Unconscious, no pulse, no breathing
**Immediate Action:**
1. Call emergency services (911/112) immediately - do not delay
2. Place person flat on firm surface
3. Begin CPR: 30 chest compressions at 100-120 per minute
4. Follow with 2 rescue breaths (if trained)
5. Continue uninterrupted CPR until AED arrives or responder arrives
6. If AED available: Turn on, attach pads to bare chest, follow voice prompts
7. Use AED and CPR together - do not stop CPR to use AED
**Recovery Position:** When breathing returns, place on side with head tilted back
**Monitor:** Check for return of pulse every 2 minutes
**DO NOT:** Stop CPR unless professional tells you to, unless person shows signs of life

### 2. Heart Attack - Chest Pain and Cardiac Symptoms
**Signs:** Chest pressure/pain (crushing, vice-like), shortness of breath, cold sweat, nausea, pain radiating to arm/jaw/back
**Immediate Action:**
1. Call emergency services immediately - describe as "chest pain - possible heart attack"
2. Have person sit down and rest in comfortable position
3. Loosen tight clothing
4. If person has prescribed nitroglycerin, help them take it
5. Chew aspirin if available and not allergic (helps prevent clot formation)
6. Monitor breathing and consciousness
7. Be ready for CPR if they become unresponsive
**DO NOT:** Delay calling for help, assume it will pass, let them refuse transport
**Time Critical:** Every minute without treatment worsens outcome

---

## RESPIRATORY EMERGENCIES

### 3. Choking - Airway Obstruction
**Signs:** Cannot speak/cry, cannot cough, difficulty breathing, blue lips/face
**Adult Choking Response (Heimlich Maneuver):**
1. Encourage coughing if they can cough
2. If cannot cough/breathe:
3. Stand behind person
4. Place fist above navel, below breastbone
5. Grasp fist with other hand
6. Quick upward thrusts into abdomen (like "J" shape)
7. Repeat 5 times
8. Alternate: 5 back blows (between shoulder blades) then 5 abdominal thrusts
9. Continue until object is expelled or person becomes unconscious
**If Unconscious:** Begin CPR immediately
**Child (1-8 years):** Use lighter pressure, same technique, adjust hand position
**Infant (under 1 year):** 5 back blows + 5 chest thrusts (not abdominal)
**Prevention:** Small objects, foods cut appropriately, no eating while laughing

### 4. Severe Asthma Attack
**Signs:** Wheezing, difficulty speaking full sentences, rapid breathing, panic, bluish lips, inability to inhale/exhale
**Immediate Action:**
1. Stop all activity
2. Have person sit upright
3. Help locate and use rescue inhaler (usually blue albuterol)
4. Inhale, pause, exhale technique - follow inhaler instructions
5. Wait 5 minutes
6. If no improvement or worsening: Call emergency services
7. Loosen tight clothing
8. Try to calm person (panic worsens symptoms)
**DO NOT:** Use expired inhalers, assume it will pass, wait too long to call help
**Warning Signs for Call 911:** Lips/nails turning blue, cannot speak, straining to breathe, altered consciousness

### 5. Anaphylaxis - Severe Allergic Reaction
**Signs:** Difficulty breathing, throat tightness, swelling (face/lips/tongue), hives, rapid weak pulse, nausea, dizziness, loss of consciousness
**Immediate Action:**
1. Call emergency services immediately
2. If person has EpiPen/auto-injector:
   - Remove from carrier tube
   - Inject into outer thigh (can go through clothing)
   - Hold in place 3-10 seconds
   - Remove and massage injection site 10 seconds
3. Have person lie down (unless vomiting - then on side)
4. Remove triggering allergen if possible
5. Be ready to give second EpiPen dose after 5-15 minutes if symptoms return
6. Stay with person, monitor breathing
**After EpiPen:** Still seek emergency medical care (symptoms may recur)
**DO NOT:** Assume one EpiPen dose is sufficient, delay calling for help, put thumb/fingers on needle end

---

## BLEEDING EMERGENCIES

### 6. Severe External Bleeding
**Signs:** Blood flowing freely, cannot be controlled with simple pressure, spurting blood, excessive soaking of dressings
**Immediate Action:**
1. Call emergency services if heavy bleeding
2. Wear gloves if available
3. Apply DIRECT PRESSURE with clean cloth - do not remove cloth
4. If blood soaks through: Layer new cloth on top (do not remove first)
5. Elevate injured area above heart if possible and no bone injury suspected
6. Maintain pressure for 5-10 minutes minimum
7. If bleeding from limb and uncontrolled:
   - Apply pressure to artery above wound (use reference for artery location)
   - As last resort: Apply tourniquet 2-3 inches above wound, tighten until bleeding stops
   - Mark tourniquet with time applied
8. If object impaled: Do NOT remove it - stabilize and get help
**Tourniquet Technique:** Wide belt/cloth, above the wound, tighten until bleeding stops, mark time
**DO NOT:** Remove first dressing, apply tourniquet unnecessarily, leave tourniquet too long (tissue damage)

### 7. Internal Bleeding - Blunt Force Trauma
**Signs:** Bruising/swelling, abdominal pain, vomiting blood, blood in stool/urine, pale skin, rapid weak pulse, dizziness, loss of consciousness
**Immediate Action:**
1. Call emergency services immediately
2. Have person lie down, legs elevated 12 inches (shock position)
3. Do not move if spinal injury suspected
4. Monitor breathing and pulse
5. Prepare for CPR
6. Do not give food or drink
7. Keep person warm with blanket
**Assessment:** Check for bruising patterns (suggests internal bleeding location)
**DO NOT:** Move person unnecessarily, give anything by mouth, massage abdomen

---

## TRAUMA EMERGENCIES

### 8. Head Injury - Concussion/Skull Fracture
**Signs:** Loss of consciousness, confusion, severe headache, bleeding from ears/nose, clear fluid from nose, unequal pupils, vomiting, seizure
**Immediate Action:**
1. If unconscious or severe symptoms: Call emergency services
2. Do NOT move head/neck (suspect spinal injury)
3. Keep airway open - position on side if necessary (use C-spine precautions)
4. Monitor consciousness and breathing
5. If conscious: Ask orientation questions (name, place, date)
6. Do not apply pressure if skull fracture suspected
7. Watch for worsening symptoms
**Post-Concussion Monitoring:** 24-48 hours observation, wake every 2 hours if slept
**Red Flags:** Unconsciousness >30 seconds, repeated vomiting, confusion worsening, seizure

### 9. Spinal Injury - Potential Spine/Neck Damage
**Signs:** Pain with movement, numbness/tingling in limbs, paralysis, loss of bladder/bowel control, mechanism (fall, accident)
**Immediate Action:**
1. Do NOT move the person
2. Call emergency services
3. Keep head/neck immobilized:
   - Hold head steady with both hands
   - If trained: Apply cervical collar
   - Prevent any twisting or bending
4. If person must be moved (danger): Log-roll technique (keep spine straight)
5. Monitor breathing and consciousness
**DO NOT:** Remove helmet if worn during injury, flex or extend neck, roll person without training

### 10. Fracture - Bone Breakage
**Signs:** Pain, swelling, deformity, inability to move, bruising, bone protrusion
**Immediate Action:**
1. Immobilize the injured area:
   - Use sling for arm fractures
   - Use splint (rolled newspaper, cardboard) for leg fractures
   - Avoid movement
2. Apply ice wrapped in cloth (15-20 minutes, 3x daily) to reduce swelling
3. Elevate if possible to reduce swelling
4. If severe or limb deformity: Call emergency services
5. Pain management: Over-the-counter pain relievers if conscious
**Splinting:** Immobilize joints above and below fracture
**DO NOT:** Try to straighten deformed limbs, apply ice directly to skin, assume fracture if unsure

---

## BURN INJURIES

### 11. Thermal Burns - Heat Damage
**Classification:**
- **1st Degree:** Red, painful, no blistering (sunburn)
- **2nd Degree:** Red, painful, blistering, may have fluid weeping
- **3rd Degree:** White/charred, may appear painless (full thickness)

**Immediate Action for All Burns:**
1. Stop heat source - remove from heat/flame
2. If clothing on fire: Drop and roll, or smother with blanket
3. Cool burn with cool (not cold) running water 10-20 minutes
4. Remove tight items (rings, bracelets) - swelling will occur
5. Cover with clean, dry cloth (do NOT apply ice directly)

**By Degree:**
- **1st Degree (small):** Cool water, pain relief, aloe vera
- **2nd Degree:** Cool water, loose bandage, pain relief, DO NOT pop blisters
- **3rd Degree/Large (>2-3 inches): Call emergency services immediately**

**Special Circumstances:**
- Chemical burn: Rinse with water 10+ minutes, remove contaminated clothing
- Electrical burn: Check consciousness, may need CPR, call emergency services
- Face/hands/genitals: Call emergency services (cosmetic/functional importance)
- Circumferential burn (around limb): Call emergency services (circulation impairment)

**DO NOT:** Apply ice directly, apply creams/ointments (except aloe vera), pop blisters, remove stuck clothing

---

## POISONING/OVERDOSE

### 12. Poisoning - Ingested/Inhaled Toxins
**Signs:** Nausea, vomiting, abdominal pain, altered consciousness, seizure, difficulty breathing, chemical smell on breath
**Immediate Action:**
1. Call Poison Control (1-800-222-1222 in US) - have bottle/substance nearby
2. If conscious: Ask what was ingested and when
3. Do NOT induce vomiting
4. If unconscious: Position on side, monitor breathing, prepare for CPR
5. Remove from contaminated area if gas inhalation
6. Follow Poison Control instructions exactly
**For Specific Toxins (examples):**
- Corrosive (lye, bleach): Do NOT vomit, sip water only
- Hydrocarbons (gasoline): Do NOT vomit (aspiration risk)
- Medications: Get specific medication name if possible
**DO NOT:** Guess treatment, delay poison control call, induce vomiting, give activated charcoal without guidance

### 13. Drug Overdose
**Signs:** Altered consciousness, shallow breathing, pinpoint pupils (opioids), rapid/irregular heartbeat, seizure, loss of responsiveness
**Immediate Action:**
1. Call emergency services immediately ("overdose" situation)
2. If opioid suspected and naloxone (Narcan) available: Administer as directed
3. Position on side if unconscious (recovery position)
4. Monitor breathing closely
5. If no breathing: Begin rescue breathing/CPR
6. Gather any substance containers for responders
7. Stay with person until help arrives
**Naloxone Use:** Inject or spray (per product), may need 2-3 doses, does not replace emergency care
**DO NOT:** Attempt to hide drug use from responders, delay calling for fear of consequences

---

## NEUROLOGICAL EMERGENCIES

### 14. Seizure/Convulsion
**Signs:** Loss of consciousness, rigid body, jerking movements, foaming mouth, loss of bladder control, post-seizure confusion
**Immediate Action:**
1. Protect from injury - move dangerous objects away
2. DO NOT restrain limbs or force objects in mouth
3. Position on side once body relaxes (prevents airway obstruction)
4. Note seizure duration
5. Monitor breathing
6. After seizure: Reorient person (may be confused), keep calm
7. If first seizure or seizure >5 minutes: Call emergency services
8. If multiple seizures: Call emergency services immediately
**Do NOT:** Put anything in mouth, force limbs down, leave person alone during recovery
**Post-Seizure State:** Person will be confused, fatigued, may not remember event - stay with them

### 15. Stroke - Sudden Neurological Deficit
**Signs (F.A.S.T. Test):**
- **F**ace: Facial drooping, smile asymmetrical
- **A**rm: Arm weakness, cannot raise both arms equally
- **S**peech: Slurred speech, difficulty understanding
- **T**ime: Note exact time symptoms started - CRITICAL

**Immediate Action:**
1. Call emergency services IMMEDIATELY - state "stroke symptoms"
2. Note exact time symptoms began
3. Have person sit/lie down safely
4. Do NOT give food or drink (swallowing may be impaired)
5. Monitor breathing and consciousness
6. Be ready for CPR
7. Keep person calm and reassured
**Time Critical:** "Time is brain" - first 3 hours critical for treatment (clot-busting medications)
**DO NOT:** Delay seeking help, assume it will improve, give medication or food

---

## ALLERGIC REACTIONS

### 16. Mild-Moderate Allergic Reaction
**Signs:** Itching, hives, mild swelling, nasal congestion, mild wheezing
**Treatment:**
1. Remove from allergen source if possible
2. Antihistamine (diphenhydramine, cetirizine) as directed on package
3. Cool compress for itching
4. Monitor for worsening
5. Seek medical evaluation if not improving in 1-2 hours
**DO NOT:** Assume mild initially - monitor for progression to anaphylaxis

---

## ABDOMINAL EMERGENCIES

### 17. Severe Abdominal Pain
**Signs:** Severe pain, rigid abdomen, vomiting, fever, shock signs
**Immediate Action:**
1. Have person lie down in comfortable position
2. Do NOT eat or drink
3. Apply heat only if told (heat may worsen some conditions)
4. Call emergency services if: Pain is severe, abdomen rigid, fever present, vomiting blood, loss of consciousness
5. Monitor vital signs
**Common Causes:** Appendicitis, perforation, internal bleeding, obstruction
**DO NOT:** Give anything to eat/drink, assume it will pass if severe, massage abdomen

---

## PSYCHOLOGICAL CRISIS

### 18. Suicide Risk Assessment
**Signs:** Talking of suicide, hopelessness, giving away possessions, mood changes, isolation, previous attempts
**Immediate Action:**
1. Take seriously - DO NOT dismiss
2. Call emergency services or suicide hotline
3. Do NOT leave person alone
4. Remove access to means (medications, weapons)
5. Listen non-judgmentally
6. Do NOT promise confidentiality if immediate danger
7. Encourage professional help
**Crisis Lines (examples):** 988 (US Suicide & Crisis Lifeline), Text "HELLO" to 741741 (Crisis Text Line)
**Follow-up:** Ensure professional mental health assessment

---

## MINOR INJURIES

### 19. Minor Cuts and Scrapes
**Care:**
1. Wash with soap and water
2. Apply antibiotic ointment
3. Cover with bandage if needed
4. Change bandage daily
**When to seek help:** Deep cuts (>1/2 inch), animal bites, dirty/rusty object, signs of infection (pus, increasing redness)

### 20. Minor Burns (1st Degree)
**Care:**
1. Cool with water
2. Apply aloe vera
3. Pain relief as needed
4. Avoid sun exposure
**Healing:** Usually 3-7 days

---

## GENERAL PROTOCOLS

### Recovery Position (Unconscious, Breathing)
1. Roll person on side
2. Tilt head back (keeps airway open)
3. Place hand under cheek for support
4. Bend top leg for stability
5. Position supports body and allows fluid drainage

### Shock Position (Consciousness, No Spinal Injury)
1. Lie flat
2. Elevate legs 12 inches
3. Keep warm with blanket
4. Monitor breathing

### CPR Compression Depth & Rate
- **Depth:** 2-2.4 inches (50-60mm)
- **Rate:** 100-120 compressions per minute
- **Continuous:** Do not stop until help arrives or person shows signs of life

---

## Emergency Contact Information
- Emergency Services: 911 (USA), 112 (EU), 999 (UK)
- Poison Control: 1-800-222-1222 (USA)
- Suicide Crisis Line: 988 (USA)
- Red Cross Training: www.redcross.org

---

**IMPORTANT DISCLAIMER**
These guidelines are for first aid training and emergency response pending professional medical care. They do NOT replace professional medical training or emergency medical services. Always call emergency services for serious conditions. Every individual case differs - respond to specific symptoms and signs as trained.

Last Updated: 2024
Sources: American Red Cross, WHO, CDC, EMS Protocols
"""

with open("content_corpus.md", "w") as f:
    f.write(content_corpus)

print("✅ Content corpus created!")
```

```python
# In your notebook cell
!pip install gradio -q

import gradio as gr
from emergency_aid_agent import EmergencyFirstAidAgent

# Initialize agent
print("🚀 Initializing Emergency First Aid Agent System...")
agent = EmergencyFirstAidAgent()
agent.setup_knowledge_base("content_corpus.md")
print("✅ Emergency First Aid Agent Ready!")

# Store chat history
chat_history = []

def process_message(message):
    """Process user message"""
    if not message:
        return "", chat_history

    # Handle reset
    if message.lower() == 'reset':
        agent.reset_conversation()
        chat_history.clear()
        return "", chat_history

    # Process emergency
    result = agent.process_input(message)
    risk = result['classification'].get('risk_level', 'UNKNOWN')
    confidence = result['classification'].get('confidence', 0)

    response = f"""**Risk Level:** {risk} (Confidence: {confidence:.1%})

**Response:**
{result['response']}

---
⚠️ For emergencies, call 911/112 immediately!"""

    chat_history.append((message, response))
    return "", chat_history

def reset():
    """Reset chat"""
    agent.reset_conversation()
    chat_history.clear()
    return chat_history

# Create interface
with gr.Blocks(title="Emergency First Aid Agent") as demo:
    gr.Markdown("""
    # 🚑 Emergency First Aid Agent
    > ⚠️ **IMPORTANT:** For guidance only. Call 911/112 for real emergencies!
    """)

    chatbot = gr.Chatbot(label="Emergency Conversation", height=400)

    with gr.Row():
        msg = gr.Textbox(label="Describe the emergency", scale=4, lines=2)
        send = gr.Button("Send 🚨", variant="primary", scale=1)

    with gr.Row():
        reset_btn = gr.Button("🔄 Reset Conversation", variant="secondary")

    # Quick examples
    gr.Markdown("### Quick Examples:")
    examples = gr.Examples(
        examples=[
            "My friend fell down the stairs and is not responding",
            "Someone is having chest pain",
            "Severe bleeding from a wound",
            "A person is unconscious"
        ],
        inputs=msg
    )

    # Wire up functions
    send.click(process_message, [msg], [msg, chatbot])
    msg.submit(process_message, [msg], [msg, chatbot])
    reset_btn.click(reset, None, chatbot)

# Launch
demo.launch(share=True)
```

```python
# In your notebook cell


import gradio as gr
import traceback
from emergency_aid_agent import EmergencyFirstAidAgent

# Initialize agent
print("🚀 Initializing Emergency First Aid Agent System...")
try:
    agent = EmergencyFirstAidAgent()
    agent.setup_knowledge_base("content_corpus.md")
    print("✅ Emergency First Aid Agent Ready!")
except Exception as e:
    print(f"❌ Error initializing agent: {e}")
    traceback.print_exc()

def chat(message, history):
    """Process emergency with error handling"""
    try:
        # Handle reset
        if message.lower() == 'reset':
            agent.reset_conversation()
            return "🔄 Reset complete. Ready for new emergency."

        # Process the input
        print(f"Processing: {message}")  # Debug print
        result = agent.process_input(message)
        print(f"Result: {result}")  # Debug print

        # Extract values safely
        classification = result.get('classification', {})
        risk = classification.get('risk_level', 'UNKNOWN')
        confidence = classification.get('confidence', 0)
        response_text = result.get('response', 'No response generated')

        return f"""**Risk Level:** {risk} (Confidence: {confidence:.1%})

**Response:**
{response_text}

---
⚠️ For emergencies, call 911/112 immediately!"""

    except Exception as e:
        error_msg = f"❌ Error: {str(e)}\n\n{traceback.format_exc()}"
        print(error_msg)
        return f"Error processing your request: {str(e)}\n\nPlease try again or type 'reset' to restart."

# Create interface
demo = gr.ChatInterface(
    fn=chat,
    title="🚑 Emergency First Aid Agent",
    description="Describe the emergency situation for first aid guidance",
    theme="soft"
)

demo.launch(share=True)
```
